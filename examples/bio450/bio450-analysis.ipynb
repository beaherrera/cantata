```python
import polars as pl
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
cmap = plt.cm.viridis
```

```python
df = pl.read_csv('out/spikes.csv')
```

```python
fg, ax = plt.subplots(figsize=(20, 20))
ax.scatter(df['time'], df['gid'], marker='|', c=df['kind'])
ax.set_xlim(0, 3000)
ax.set_ylim(0, 550)
ax.set_xlabel('t/ms')
ax.set_ylabel('Cell')

custom_lines = [Line2D([0], [0], color=cmap(0/2), lw=4),
                Line2D([0], [0], color=cmap(1/2), lw=4),
                Line2D([0], [0], color=cmap(2/2), lw=4)]
ax.legend(custom_lines, ['Cable', 'LIF', 'Virtual'])
```
