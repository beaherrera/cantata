```python
import polars as pl
import matplotlib.pyplot as plt
```

```python
df = pl.read_csv('out/spikes.csv')
```

```python
df.head(100)
```

```python
fg, ax = plt.subplots(figsize=(20, 20))
ax.scatter(df['time:'], 4*df['gid'], marker='|')
ax.set_xlim(0, 3000)
ax.set_ylim(0, 2500)
ax.set_xlabel('t/ms')
ax.set_yticks([])
ax.set_ylabel('Cell')
```

```python
fg.savefig('bio-450-raster.pdf')
fg.savefig('bio-450-raster.png')
```
